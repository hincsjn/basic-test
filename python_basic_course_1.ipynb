##### 2. Операции с числами

Для вывода примеров на экран мы воспользуемся функцией *print()*.<br>
О ней мы поподробнее поговорим позже, а пока будем просто записывать<br>
то, что хотим вывести на экран в скобки к данной функции

```python
# Сложение
print(14 + 5) 
# Произведение
print(14 * 5)
# Вычитание
print(14 - 5)
# Возведение в степень
print(14 ** 2)
```

```python
# Деление (обычное)
print(14 / 5)
# целочисленное деление
print(14 // 5)
# получение остатка от деления
print(14 % 5)
# ВНИМАНИЕ! Остаток и дробная часть - разные вещи!
```

Как найти остаток от деления? <br>
Находим наибольшее число до «14», которое делится на «5» без остатка. Это «10».<br>
10 : 5 = 2<br>
Вычитаем из делимого найденное число из пункта «1».<br>
14 − 10 = 4

#####    3.Переменные

Программисты, при написании программ часто обращаются к такому понятию, как переменные. Суть в следующем: практически всегда, когда мы пишем код, нам требуется как-то получать, хранить, обрабатывать и возвращать данные. Это могут быть очки здоровья игрока, пароль от аккаунта в соц. Сети или что-то другое. Чтобы хранить подобные данные, их нужно записать в память компьютера, а, чтобы не потерять их, мы кладем их в переменные. Все это можно представить следующим образом: получаем имя от пользователя, кладем это имя в коробку (переменную), а затем эту коробку засовываем в шкаф памяти компьютера. Когда нам потребуется что-то сделать с этой информацией, мы просто назовем имя коробки (переменной) и тем самым получим информацию, которая в ней находится. Мы ее можем изменить, вывести на экран или произвести другие операции. Имена можно задать любые, главное, чтобы они были понятны нам. Нельзя начинать имя переменной с символа или числа и используем только латинские буквы.

```python
# Создание (инициирование) переменной
name = 'Коля'
age = 15
# Вывод значений, записанных в переменных
print(name)
print(age)
```

Это так называемая операция присвоения. В ней создается переменная которой <br>
присваивается значение name — ссылка на строку «Коля». Вообще принято <br>
говорить, что переменным присваиваются значения. Обрати внимание, что названия <br>
переменных совпадают со значением данных, записанных в них. Но это сделанно <br>
только ради удобства, на самом деле назвать переменные можно иначе.

```python
# точно так же можно изменить, записанные в переменной данные
# вернее будет сказать, что они перезаписываются
name = 'Миша'
age = 20
print(name)
print(age)
```

##### 4.Типы данных

В python мы будем работать с разными типами данных: <br>
str - строки (str от английского string - строка), при работе со строками используем кавычки (st = “Привет всем!”)<br>
int - целые числа (int от английского integer - целый)<br>
float - числа с плавающей точкой (дробные числа) (float - плавающий)<br>

Порой происходят ситуации, когда требуется перевести один тип данных, в другой. Для преобразования <br>
одного типа данных в другой используем функции, названия которых совпадает с названием типа, в который <br>
хотим перевести переменную. А для того, чтобы узнать, какого типа конкретная переменная, нужно всего-то <br>
воспользоваться функцией *type()*

```python
# Создаем переменные
num1 = 10
num2 = 10.0
string = 'Привет!'
# Выводим типы переменных на экран
print(type(num1))
print(type(num2))
print(type(string))
```

Заметим, что тип переменной num1 - integer (целое число),<br> 
Чтобы перевести ее в тип float, к примеру, воспользуемся<br>
Соответствующей функцией *float()*

```python
num1 = float(num1)
print(num1)
print(type(num1))
```

**Другой пример**

```python
a = '50'
print('Переменная а является строкой:')
print(a)
print(type(a))
a = int(a)
print()
print('После перевода, тип переменной а изменился:')
print(a)
print(type(a))
```

#### Функция *input()* 

Очень часто полезные программы получают данные, хранят их, обрабатывают и возвращают.<br>
Сегодня мы познакомимся с функцией *input()*. Эта функция позволяет<br>
нам получать значения от пользователя и сохранять их в переменной уже после запуска программы.<br>
Пример:

```python
name = input()
print('Привет,', name)
```

Давай теперь попробуем написать программу, которая складывает два числа

```python
a = input()
b = input()
print(a + b)
```

Очень странно, почему-то получилось 35, а не 15. <br>
**Важно!** Функция *input()* воспринимает все введенные данные в виде типа str (строка).<br> 
Поэтому, если мы введем число и умножим его на другое, то получим не то, что хотели:

```python
number = input()
print(number * 4)
print(type(number))
```

Поэтому, если мы хотим получить от пользователя число, то нам его нужно привести к числовому типу данных:

```python
number = input()
number = int(number)
print(number * 4)
print(type(number))
```

Но не стоит забывать и о типе данных float

```python
number = input()
number = float(number)
print(number * 4)
print(type(number))
```

#### Модуль turtle 

Модуль (библиотека) - это папка, то есть набор функций, которые расширяют возможности языка. <br>
Сегодня мы начнем знакомство с модуля turtle, который позволяет нам рисовать на python

Для работы с любым модулем сначала нужно его импортировать, то есть добавить в программу. <br>
Воспользуемся командой *import* и запишем имя модуля

```python
import turtle
```

```python
# специальная версия модуля для работы вjupyter notebook
# from mobilechelonian import Turtle
# t = Turtle()
```

Теперь, нужно создать объект черепашки, которая будет рисовать. <br>
Создадим переменную *t*, в которой сохраним объект черепашки, обратившись к имени модуля <br>
Подробнее рассмотрим тему объекта в программировании в целом и в python в частности чуть позже

```python
t = turtle.Turtle()
```

Каждый раз, когда мы хотим произвести какую-то операцию с черепашкой, будем обращаться к ней через переменную и точку.
Для начала, зададим "настройки" нашей черепашки:

```python
# внешний вид черепашки
t.shape('turtle')
# скорость("скорость")
t.speed(10)
# цвет("цвет")
t.color('purple')
# ширина линии(ширина в пикселях)
t.width(3)
```

Теперь поговорим о базовых функциях данной библиотеки

```python
# двигаться вперед на 100 пикселей
t.forward(100)
# повернуться направо/налево на 90 градусов
t.right(90)
t.left(90)
# двигаться назад на 100 пикселей
t.backward(100)
# поднять перо (перестать оставлять след)
t.penup()
# опустить перо (продолжить оставлять след)
t.pendown()
# очистить холс (поле для рисования)
t.clear()
# переместить черепашку на координаты x=40, y=60
# (центр экрана - координаты х=0, у=0)
t.goto(40, 60)
```

##### Условная конструкция

Программы, которые выполняются последовательно, построчно чаще всего малополезны. Поэтому существуют оператор ветвления, или, по-другому, оператор принятия решений. Называется он if-elif-else (если - иначе если - иначе)

Сначала записывается часть if с условным выражением, далее могут следовать одна или более <br>
необязательных частей elif, и, наконец, необязательная часть else. Общая форма записи условной <br>
инструкции if выглядит следующим образом:

```python
if 'условие1':
    "выражение1"
elif "условие2":
    "выражение2"
else:
    "выражение3"
```

##### 3. Слайсы

Но получать из списка можно не только отдельные элементы, но и сразу несколько оных. Получение сразу нескольких элементов из списка называется срезом. Для того, чтобы получить срез нужно указать в квадратных скобках начало и конец среза. Не забывайте, что при работе со срезом мы используем индексы элементов.

```python
nums = [5, 2, 6, 4, 10, 3, 2, 5, 7, 11]
# получить все элементы, кроме первого
print(nums[1:])
# получить элементы с индексами от 2 до 5(не включительно)
print(nums[2:5])
# получить все элементы до элемента с индексом 7
print(nums[:7])
```
